### 1.1. An overview of the end-to-end pipeline

Um pipeline de Machine Learning (ML) é uma sequência de etapas para conduzir um projeto de ML. Essas etapas seguem:

- **Definicão do problema e coleta de dados:** Define objetivo do problema em termo de negócios como um problema de ML e colete os dados de que você precisa.

- **Pré-processamento de dados e engenharia de recursos:** Processar os dados em um formato adequado que
pode ser inserido nos algoritmos de ML.Assim como,selecionar ou gerar recursos relacionados
para a saída de destino para melhorar o desempenho dos algoritmos.Além disso, essa etapa geralmente é realizada explorando primeiro o conjunto de dados para ter uma noção de suas características. As operações devem acomodar os algoritmos de ML específicos que você está considerando.

- **Seleção de algoritmo de ML:** Selecione algoritmos de ML apropriados para a tarefa que você
gostaria de testar, com base em seu conhecimento prévio do problema e em sua
experiência.

- **Treinamento e avaliação do modelo:** Aplique o algoritmo (ou algoritmos) de ML selecionado
para treinar um modelo de ML com seus dados de treinamento e avaliar seu desempenho em
o conjunto de dados de validação.

- **Ajuste de hiperparâmetros:** Tente alcançar um melhor desempenho iterativamente
ajustando os hiperparâmetros do modelo.

- **Implantação de serviços e monitoramento de modelos:** Implante a solução final de ML e monitore seu desempenho para que você possa manter e melhorar o pipeline continuamente.

### 2.2.1.1. SMA and  MSD

```python
# Function SMA and MSD
def sma_and_msd(df):

    # Copy of dataframe
    df_copy = df.copy()

    # Cria a variável com o retorno (percentage change in total sales per hour from the previous hour )
    # This will our variable target
    df_copy["perc_return"] = df_copy["total_sales_per_hour"].pct_change(1) * 100

    # Shift das colunas op sales toys
    df_copy["sa"]  = df_copy["sales_amount"].shift(1) 
    df_copy["sd"]  = df_copy["sales_discount"].shift(1)
    df_copy["sd"]  = df_copy["final_sales_price"].shift(1)
    df_copy["tsd"] = df_copy["total_sales_per_day"].shift(1)
    df_copy["tsh"] = df_copy["total_sales_per_hour"].shift(1)
    df_copy["avsd"]= df_copy["avg_sales_per_day"].shift(1)
    df_copy["avsh"]= df_copy["avg_sales_per_hour"].shift(1)

    # Calculando os OKR's

    # Simple Moving Average (SMA)
    df_copy["SMA_3"] = df_copy[["total_sales_per_hour"]].rolling(3).mean().shift(1)
    df_copy["SMA_5"] = df_copy[["total_sales_per_hour"]].rolling(5).mean().shift(1)

    # Moving Standard Deviation (MSD) - Volatilidade
    df_copy["MSD_3"] = df_copy["perc_return"].rolling(3).std().shift(1)
    df_copy["MSD_5"] = df_copy["perc_return"].rolling(5).std().shift(1)

    return df_copy.dropna()

# Function call
df_new = sma_and_msd(df_dropna)
```

### 2.3.9. Teste Dickey-Fuller Aumentado

O teste de estacionriedade usual, proposto Dickey-Fuller, em 1979, conhecido **Teste de Dickey-Fuller aumentado ou Teste ADF (do acrônimo em inglês Augmented Dickey-Fuller)** é um teste de raiz unitária em séries temporais. Ele é uma versão aumentada do Teste de Dickey-Fuller, é aplicado a modelos mais complicados de séries temporais. 


A estatística ADF do  teste assume uma hipótese nula de raiz unitária na série temporal. Dependendo dos resultados do teste, essa hipótese nula pode ser rejeitada no nível de significância especificado, o que significa que a presença de um teste de raiz unitária pode ser rejeitada no nível de significância fornecido.

O nome do teste faz referência aos estatísticos D. A. Dickey e W. A. Fuller.


A intuição por trás do teste é que, se a série for integrada, o nível de atraso da série y (t-1) não fornecerá informações relevantes na previsão da mudança em y (t).


| Hipótese| Descrição do teste|
|---|:-------------|
| **H0 (Hipótese Nula):** |A série temporal **não é estacionária.**|
| **H1 (Hipótese Alternativa):**|A série temporal **é estacionária.**|

Um valor-p menor que 0.05 significa que podemos rejeitar a hipótese nula de que os dados da série temporal **não são estacionários.** Em outras palavras, a série temporal **é estacionária.**

Um valor-p maior que 0.05 significa que **não rejeitamos** a hipótese nula. Em poucas palavras, a série temporal **não é estacionária.**

### 2.3.10. Applicability normality Test (Shapiro-Wilks Test)

Quando tomanos o nome teste de normalidade, ficamos confusos sobre como encontrar o melhor teste para saber se a variável segue uma distribuição normal. Existe muitos testes estatísticos, incluindo de correlação, regressão, teste t e análise de variância (ANOVA), assumem certas características sobre os dados, ou melhor, poussi estimativas de parâmetros. Assim, de fato os testes requerem que os dados segue uma **distribuição normal**. Esses testes são conhecidos de **testes paramétricos**, porque sua validade depende da distribuição dos dados.

Ainda mais, antes de usar testes paramétricos devemos executar alguns testes preliminares para garantir que as suposições do teste sejam atendidas. Nos casos em que as suposições são violadas, testes não paramétricos são recomendados. É importante destacar que **os testes não paramétricos geralmente são menos poderosos do que os testes paramétricos correspondentes quando vigora a suposição de normalidade. Assim, é menos 
provável que você rejeite a hipótese nula quando ela é falsa se os dados vierem da distribuição 
normal.**

> Os métodos não paramétricos são úteis quando a suposição de normalidade não se 
sustenta e seu tamanho da amostra é pequeno.
 
É possível usar um **teste de significância** comparando a distribuição da amostra com a normal, com interesse de investigar se os dados mostram ou não um desvio śerio  de normalidade.

Existem diversos métodos para o **teste de normalidade, como o teste de Shapiro-Wilk.** O método de Shapiro-Wilk é amplamente recomendado para teste de normalidade. É baseado na correlação entre os dados e as pontuações normais correspondentes, e é ele que utilizaremos em nossas análises a seguir. Este teste é uns dos mais popular para testar a normalidade dos dados.

As hipóteses deste teste incluem:

| Hipótese| Descrição do teste|
|---|:-------------|
| **H0 (Hipótese Nula):** |Os dados são normalmente distribuídos. Em outras palavras, as amostras vêm de uma distribuição normal. |
| **H1 (Hipótese Alternativa):**|Os dados não são normalmente distribuídos. Caso contrário, a amostra não vem de uma distribuição normal.|



Em poucas palavras, o teste de Shapiro-Wilk calcula se uma amostra aleatória de dados, segue uma distribuição normal. Quando o valor de **p é menor ou igual a 0.05 (assumindo um nível de confiança de 95%), os dados não são normais.** Portanto, se este teste falhar, você pode afirmar com 95% de confiança que seus dados não se encaixam  próximo de uma distribuição normal

- Se o valor-p for **maior que 0.05** não rejeitamos a hipótese nula e podemos assumir a normalidade dos dados.
- Se o valor-p for **menor que 0.05** rejeitamos a hipótese nula e **não** podemos assumir a normalidade dos dados.

### 5.1 Regression loss function

Em redes neurais, **uma função de perda é usada na otimização dos valores de peso de um modelo.** Uma função de perda realiza o mapeamento das decisões para seus pesos associados. A função de perda também é conhecida como a **função de erro.**

**Uma função de custo geralmente é uma perda média para todo o conjunto de dados usado para treinar o modelo.** Mas uma função de perda geralmente é **usada para um único exemplo no conjunto de dados de treinamento**. Embora a função de perda e as funções de custo sejam intercambiáveis ou usadas como sinônimos uma da outra, elas são, na verdade, diferentes.


Em outras palavras, as funções de perda quantificam o quão perto uma determinada rede neural esta do ideal para o qual está sendo treinada. A ideia é simples, calculamos uma métrica com base no **erro que observamos nas previsões da rede.** **Em seguida, agregamos esses erros em todo o conjunto de dados e calculamos a média deles e agora temos um único número representativo de quão próximo a rede neural está em seu ideal**


- Procurar este estado ideal equivale a encontrar os parâmetros (pesos e vieses) que irão minimizar a “perda” incorrida com os erros. Dessa forma, as funções de perda ajudam a reformular o treinamento de redes neurais como um problema de otimização. Na maioria dos casos, esses parâmetros não podem ser resolvidos analiticamente, mas, na maioria das vezes, podem ser bem aproximados com algoritmos de otimização iterativos, como gradiente descendente.

Vamos recapitular alguns pontos-chave do capítulo:

**Função de perda:** Avalia o quão bem ela classifica (minimiza o erro) em cada etapa

**Função de otimização:** Orienta-o para os pontos de menor erro

> A função de perda resume a diferença entre desejado e previsto, sejam eles vetores, em um único número.

### 5.2. A common-sense, non-machine-learning baseline

**The Mean absolute error (MAE)**, também conhecido como perda L1, é uma das funções de perda mais simples e, é uma métrica de avaliação fácil de entender em modelos de regressão. Calculamos o resíduo para cada ponto de dados, tomando apenas o valor absoluto de cada um para que os resíduos negativos e positivos não se anulem. Em seguida, calculamos a média de todos esses resíduos. Assim como, MAE mede apenas a magnitude dos erros e não se preocupa com sua direção. Quanto menor o MAE, maior a precisão de um modelo.

Em outras palavras, MAE mede a magnitude média dos erros em um conjunto de previsões, **sem considerar sua direção**. Partindo desse pressuposto que, **o MAE é a média sobre a amostra da verificação dos valores absolutos das diferenças entre a previsão e a observação correspondente.** Assim , pode-se então afirma que o MAE é uma pontuação linear, o que significa que todas as diferenças individuais são ponderadas igualmente na média. Saiba mais  <a href="https://resources.eumetrain.org/data/4/451/english/msg/ver_cont_var/uos3/uos3_ko1.htm" target="bank">Mean Absolute Error (MAE)</a>, a equação formal é mostrada abaixo.

No entanto, se o valor absoluto não for obtido (os sinais dos erros não forem removidos), o erro médio torna-se o erro de viés médio (MBE) e geralmente é destinado a medir o viés médio do modelo. O MBE pode transmitir informações úteis, mas deve ser interpretado com cautela porque os erros positivos e negativos serão anulados.Consulte o link para sabe mais sobre este detalhes <a href="https://medium.com/human-in-a-machine-world/mae-and-rmse-which-metric-is-better-e60ac3bde13d" target="_bank">MAE and RMSE — Which Metric is Better?</a>

**Lembrete:**

**O  módulo  (ou valor absoluto) de um número sempre retorna *seu valor positivo*. Rigorosamente:**
    
    se A => 0, então  |A| =  A
    se A <  0, então  |A| = -A
    
Desse modo:

- **|3|** = 3

- **|-0,273|** = 0,273

**Termos-chave:**

**Erro Médio Absoluto (sigla MAE, em inglês)** também chamado de **Desvio Médio Absoluto**. É a soma da diferença absoluta entre previsões e valores reais. Dessa maneira, formenta a ideia de quão erradas estão nossas previsões. Portanto, valor igual a 0 indica que não há erro, sendo a previsão perfeita (o que é difícil de acontecer).

De modo menos formal, quando subtraímos os valores previstos dos valores reais, obtendo os erros, somamos os valores absolutos desses erros e obtemos sua **média**. Essa métrica dá uma noção do erro geral para cada previsão do modelo, quanto menor (mais próximo de 0) melhor.

### 6.7. Creating a tuner for tuning Keras models

Um sintonizador contém um método de pesquisa e organiza o **treinamento e a avaliação dos pipelines selecionados durante o processo de pesquisa.** Diferentes métodos de pesquisa podem ser escolhidos alterando o oráculo no KerasTuner.

Embora o KerasTuner forneça uma maneira universal de personalizar o espaço de pesquisa definindo a **função build()** com a ajuda do contêiner hp, treinar, avaliar, salvar e carregar esses modelos ainda pode exigir tratamentos diferentes. Isso pode ser resolvido definindo seus próprios sintonizadores. A razão é que um sintonizador no KerasTuner controla o treinamento e a avaliação dos modelos instanciados durante o processo de pesquisa.


Nesta etapa do projeto, vamos criar um sintonizador personalizado para ajustar os modelos DL implementados com o TensorFlow Keras. Usamos um sintonizador embutido para ajustar o modelo do keras: o sintonizador **RandomSearch**, que é codificado com o método de pesquisa aleatória
(oráculo é o **oráculo de busca aleatória e não pode ser alterado**). No entanto, agora criamos um sintonizador personalizado que pode selecionar entre diferentes métodos de pesquisa, criamos um tuner chamado **DeepTuner estendendo a classe base do tuner.** Se a função de inicialização e a função de pesquisa não tiverem nenhuma operação especializada em comparação com as do sintonizador de base, podemos ignorá-las. Assim, implementamos apenas três funções para o DeepTuner aqui: as funções **run_trial(), save_model() e load_model()**, segundo (HAIFENG JIN; QINGQUAN SONG; XIA “BEN” HU, 2022).


Além disso, ampliando a função run_trial(), ainda podemos construir o modelo Keras chamando a **função build() do hipermodelo.** Em seguida, chamamos a **função fit()** do modelo Keras instanciado para treiná-lo. Notavelmente, o treinamento dos modelos de DL pode exigir hiperparâmetros extras, como tamanho do **lote** e número de **épocas**, para ajudar a controlar o **algoritmo de otimização**. Podemos passar esses hiperparâmetros nos argumentos **fit_kwargs. Ainda mais, depois que o modelo for treinado, podemos avaliá-lo com os dados de validação e usar os resultados da avaliação para **atualizar o oráculo.** Especificamente, você pode questionar como atualizar o oráculo quando temos múltiplas métricas de avaliação. Por exemplo, por padrão, a **função evaluate()** do Keras retornará o valor da perda de avaliação. Uma maneira direta de resolver isso é salvar a avaliação de todas as métricas em um **dicionário** com os nomes das métricas como chaves e alimentar todas elas no **oráculo**. **Ao inicializar o sintonizador, podemos informar ao sintonizador qual métrica específica queremos usar para comparar os modelos e atualizar o oráculo. Como o TensorFlow Keras fornece métodos para salvar e carregar os modelos, podemos adotar esses métodos para implementar as funções save_model() e load_model()** (HAIFENG JIN; QINGQUAN SONG; XIA “BEN” HU, 2022).

Importante ainda destacar que substituir a classe Tuner nos dá controle total sobre o espaço de pesquisa, construção de modelo, treinamento, salvamento e processo de avaliação.


```python
# Creating a tuner called DeepTuner by extending the base tuner class
import keras_tuner as kt
class DeepTuner(kt.Tuner):
    def run_trial(self, trial, X, y, validation_data, **fit_kwargs):
        model = self.hypermodel.build(trial.hyperparameters)
        model.fit(X, y,
                  batch_size=trial.hyperparameters.Choice("batch_size", [32, 64]),
                  **fit_kwargs
                  )
        X_val, y_val = validation_data
        eval_scores  = model.evaluate(X_val, y_val)
        self.save_model(trial.trial_id, model)
        return {
            name: value for name, value in zip(model.metrics_names, eval_scores)
        }
    def save_model(self, trial_id, model, step=0):
        fname = os.path.join(self.get_trial_dir(trial_id), "models")
        model.save(fname)
    def load_model(self, trial):
        fname = os.path.join(self.get_trial_dir(trial.trial_id), "models")
        model = tf.keras.models.load_model(fname)
        return model
```

Na próxima seção, criaremos uma função para **instanciar o sintonizador personalizado** para ajustar o nosso modelo. **Como mostra a função que criamos para espaço de pesquisa de MLPs, ou seja, função build_model() e usamos para inicializar um objeto DeepTuner.** Acresenta-se que, precisamos selecionamos o oráculo, que é o sintonizador RandomSearch não pode fazer. **O objetivo é especificado como erro médio absoluto (MAE), que será a métrica para comparar os modelos (usando kt.Objective) e usada na função run_trial() para atualizar o oráculo. Ao chamar a função de pesquisa, podemos executar a pesquisa do processo.** O número de épocas será passado da função de pesquisa para o
run_trial() (através do **fit_kwargs) para controlar as épocas de treinamento de cada MLP selecionado.


### 6.8. Creating the search space 

Em primeiro lugar, precisamos de uma função build(), ou seja, esta **função é a parte central do AutoML**, ao qual possa gerar um modelo MLP (para este cenário) e definir o espaço de busca para os hiperparâmetros do AutoML com um espaço de busca totalmente customizado com base nos requisitos deste projeto.

Em outras palavras, além da arquitetura da rede, também precisamos criar o nó de entrada, configurar a camada de regressão de saída, selecionar o algoritmo de otimização e compilar o modelo para treinamento. Resumindo, **nesta função devemos criar todos os componentes do pipeline de DL na função build().**



Além do mais, os autores ressalta que, a construção do espaço de busca geralmente desempenha um papel vital no sucesso do AutoML. Assim como, um bom espaço de busca pode ajudá-lo a descobrir um pipeline promissor com menos tempo de pesquisa. Projetar um bom espaço de pesquisa pode ser ainda mais importante do que projetar um bom algoritmo de pesquisa, porque ele fornece restrições suncitas para acelerar o processo de pesquisa. No entanto, muitas vezes requer conhecimento prévio e compreensão dos modelos. Se você não tem conhecimento prévio para desenvolver, você pode começar com umgrande espaço de pesquisa e diminuir gradualmente seu tamanho com uma abordagem de tentativa e erro. Essa ideia também motiva alguns algoritmos avançados de AutoML que visam adaptar gradualmente o espaço de pesquisa ou reduzi-lo a uma região mais fina.

Em seguida, implementamos a função build(). Ela tem duas funções: **definir o espaço de busca para os hiperparâmetros que queremos ajustar** e construir um **MLP em cada tentativa durante o processo de pesquisa sempre que for chamado pelo algoritmo de pesquisa**.

```python
# Creates a search space for tuning MLPs
def build_regressor(hp):
    model = tf.keras.models.Sequential()
    model.add(tf.keras.Input(shape=(5,)))
    for i in range(hp.Int("num_layers", min_value=1, max_value=4)):
        model.add(
            tf.keras.layers.Dense(
                hp.Int("units_{i}".format(i=i), min_value=32, max_value=512, step=32),
                hp.Choice('activation',['relu','selu']),
            )
        )
    model.add(tf.keras.layers.Dense(1)) 
    optimizer_name = hp.Choice("optimizer", ["adam", "nadam"])
    learning_rate  = hp.Float("learning_rate", min_value=1e-3, max_value=0.1, sampling="log")
    if optimizer_name == "adam":
        optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    else:
        optimizer = tf.keras.optimizers.Nadam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, 
                  loss=tf.keras.losses.Huber(), 
                  metrics=["mae", "mape"]) 
    return model
```

Depois de definir **o espaço de busca**, ainda precisamos de um **método de busca** para ajudar a **conduzir o processo de pesquisa.** Lembre-se que, ao criar um pipeline do AutoML,
envolva os blocos em um objeto AutoModel e, em seguida, defina seu argumento do sintonizador (tuner) (ou use o
valor padrão, 'greedy') **para selecionar um método de pesquisa para ajuste**.
**Um sintonizador não apenas especifica um método de busca, mas também ajuda a agendar o treinamento e
avaliação dos modelos selecionados em cada tentativa de busca para garantir a busca sequencial
do processo, ao qual pode prosseguir sem problemas**. O nome do sintonizador está de acordo com o método de busca que queremos usar. Por exemplo, podemos usar o método de pesquisa aleatória escolhendo o sintonizador RandomSearch.

### 7. Getting started with a random search method

Nesta seção, apresentaremos como criar um método de pesquisa aleatória com o KerasTuner para explorar o **espaço de pesquisa e encontrar melhores hiperparâmetros.** O método de **RandomSearch** é uma das maneiras mais simples e tradicionais de fazer o **ajuste de hiperparâmetros no AutoML.** O método de **RandomSearch** explora aleatoriamente as combinações de hiperparâmetros no espaço de pesquisa. Essa abordagem tem se mostrado empiricamente mais poderosa do que o método do GridSearch  na maioria dos casos.


- Why is random search often better than grid search?

Você pode encontrar mais detalhes no artigo "Random Search for Hyper-Parameter Optimization" de James Bergstra e Yoshua Bengio (www.jmlr.org/papers/v13/bergstra12a.html)

O método de busca no KerasTuner é implementado como um objeto Oracle que pode ser chamado pelos sintonizadores.

Vale ressaltar que os hiperparâmetros do método de pesquisa não estão contidos no **espaço de pesquisa**, e precisamos sintonizá-los nós mesmos. Eles são considerados hiper-hiperparâmetros, que são os hiperparâmetros usados para controlar o processo de ajuste de hiperparâmetros.

```python
# Use build-in random search algorithm to tune models
def random_tuner(X_train, y_train, X_valid, y_valid,
                 max_trials, exec_per_trial,
                 directory,name, 
                 patience=None, epochs=None,seed=42):
    
    rt = DeepTuner(
        oracle=kt.oracles.RandomSearch(objective=kt.Objective("mae", "min"),
                                        max_trials=max_trials,
                                        seed=seed,
        ),
        hypermodel=build_regressor, 
        overwrite=True,
        executions_per_trial=exec_per_trial,
        directory=directory,
        project_name=name,
    )
    early_stopping_cb = tf.keras.callbacks.EarlyStopping(monitor="mae",mode="auto",
                                                         patience=patience,
                                                         restore_best_weights=True,
        )
    rt.search(X_train, y_train,validation_data=(X_valid, y_valid),
              callbacks=early_stopping_cb,
              epochs=epochs,verbose=2
    )
    return rt
```

Com efeito da criação desta função, fornecemos o espaço de pesquisa (a função build()) durante a inicialização
e **definir o número de modelos que queremos explorar durante o processo de pesquisa (max_
tentativa = 50).** O argumento objetivo especifica a métrica (ou função de perda) que queremos usar para comparar os modelos e otimizar o método de busca (RandomSearch method, não há nada para otimizar). **Neste exemplo, em cada tentativa, o tuner RandomSearch construirá um modelo com base no valor das unidades selecionadas aleatoriamente, treinará o
modelo e avaliará .** Ele retornará o melhor modelo com base no objetivo que definimos,
que é o erro médio absoluto (MAE) no conjunto de validação (indicado pelo val_
prefixo). Este conjunto de validação, se não for especificado, será separado aleatoriamente do conjunto de treinamento. **Podemos especificar quantas vezes queremos que cada modelo seja treinado e avaliado, para reduzir a aleatoriedade no processo de avaliação, definindo no argumento executions_

### 7.1. Evaluate the best discovered model and Predict

> Metrics: Estatísticas para monitorar durante o processo de treinamento e teste, como a precisão de uma tarefa de regressão (este cenário). Estas não afetarão o processo de treinamento, mas serão calculadas e registradas tanto para o treinamento quanto para o conjunto de validação. Esses valores são usados como informações auxiliares para analisar e ajustar o algoritmo projetado.

### 7.2. Creating a function to evaluate the best discovered model

```python
def evaluate_and_predict(tuner,X_test, y_test, idx_x_test,idx_y_test):
    
    # Get the optimal hyperparameters
    best_hps= tuner.get_best_hyperparameters(1)[0]
    
    # Recover to the best model
    best_model = tuner.get_best_models(1)[0]
    
    # Prints a summary of the architecture of the model
    best_model.summary()
    
    # Evaluate best pipeline
    y_pred = best_model.evaluate(X_test, y_test)
    #print(best_model.metrics_names)
    
    #Evaluate the best discovered model
    evaluation_rules.regression(best_model, X_test, y_test) 

    
    # The inverse transformation
    y_pred_test = pd.DataFrame(scaler_target.inverse_transform(best_model.predict(X_test)), 
                               columns=['predictions'], index=idx_x_test.index)
    tspd = pd.DataFrame(scaler_target.inverse_transform(y_test), 
                        columns=['total_sales_per_hour'], index=idx_y_test.index)
    
    # Compare 
    predict = tspd.join(y_pred_test)
    
    # Calcule a difference
    pd.options.display.float_format = "{:,.2f}".format
    predict['Diff'] =  abs(predict.total_sales_per_hour - predict.predictions)

    # Return dataframe with last 5 predictions
    return predict.tail(5)
```

### 7.3. Retrain the model

Por fim, construiremos um modelo usando os hiperparâmetros ideais antes de chamar a função fit para treinar o modelo.

```python
X_train_all= np.concatenate((X_train_min_max, X_valid_min_max))
y_train_all= np.concatenate((y_train_min_max, y_valid_min_max))

X_test_all= np.concatenate((X_test_min_max, X_valid_min_max))
y_test_all= np.concatenate((y_train_min_max,y_valid_min_max))

print(f'Dataset Original:{X_train_all.shape, y_train_all.shape, X_test_all.shape, y_test_all.shape}')
print(f'Dasetd Valid:    {X_train.shape, y_train.shape, X_valid.shape, y_valid.shape}')
```

```python
# Creating an function for Retrain the model. 
def retrain_model(tuner,
                  X_train,y_train,X_valid,y_valid,
                  X_test,y_test,idx_x_test,idx_y_test,
                  name,
                  epochs=None, patience=None
                 ):
    
    # Get the optimal hyperparameters
    best_hps= tuner.get_best_hyperparameters(1)[0]
    
    # Using HyperModel also allows you to retrain the best model by yourself.
    model = tuner.hypermodel.build(best_hps)
    # Retrain the model, in other words,  Train & eval model
    model.fit(X_train,y_train,
              validation_data=(X_valid,y_valid),
              epochs= epochs,
              callbacks=[tf.keras.callbacks.EarlyStopping(patience=patience)],
              verbose=0,
    )
    
    # Save  model
    model.save("models/"f"{name}",overwrite=True)
    
    # Evaluate the model
    Y_pred = model.evaluate(X_test,y_test)
    
    #Evaluate the model
    evaluation_rules.regression(model, X_test, y_test) 
    
    # The inverse transformation
    y_pred_test = pd.DataFrame(scaler_target.inverse_transform(model.predict(X_test)), 
                               columns=['predictions'], index=idx_x_test.index)
    
    tspd = pd.DataFrame(scaler_target.inverse_transform(y_test), 
                        columns=['total_sales_per_hour'], index=idx_y_test.index)
    
    # Compare 
    predict = tspd.join(y_pred_test)
    
    # Calcule a difference
    pd.options.display.float_format = "{:,.2f}".format
    predict['Diff'] =  abs(predict.total_sales_per_hour - predict.predictions)
    # Return dataframe with last 5 predictions
    return predict.tail(5)
```

### 9.1 Plot search curves

```python
# Creating functions for plot search curves
def plot_curve(x, y, xlabel, ylabel, title):
    plt.figure(figsize=(18,10))
    plt.plot(x, y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()


def plot_curves(
    x, ys, xlabel, ylabel, title, ymin, ymax, legend, markers, linestyles, markevery=1
):
    plt.figure(figsize=(18,10))
    for i, y in enumerate(ys):
        plt.plot(x, y, marker=markers[i], linestyle=linestyles[i], markevery=markevery)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.ylim(ymin, ymax)
    plt.legend(legend)
    plt.show()
```

Em primeiro lugar, para capturar e mostra como é o processo de pesquisa, extraímos o **desempenho da avaliação de todos os modelos pesquisados e os plotamos em ordem.** Os modelos são registrados na ordem em que as tentativas terminaram como uma lista no atributo **end_order do oráculo**, que é **rt.oracle.end_order e bt.oracle.end_order** neste projeto (em nosso exemplo). A ordem final das tentativas é igual à ordem inicial, porque não estamos fazendo tentativas paralelas neste caso. O código para plotar a curva de pesquisa é mostrado a seguir.

Ao contrário do método Random Search, o desempenho dos modelos buscados pela Bayesian optimization se estabiliza gradualmente à medida que o processo de busca continua, ou seja, ela tem uma convergência mais apropriada a partir de 20 pipelines de teste. Isso ocorre porque a Bayesian optimization leva em conta as informações históricas e pode alavancar a exploração para pesquisa, de modo que os modelos descobertos posteriormente provavelmente alcançarão desempenho comparável ou até melhor do que os descobertos anteriormente. É importante analisar que **Random Search não pode levar em conta avaliações históricas, os modelos descobertos posteriormente não se beneficiam e tendem a não ser melhores que os resultados anteriores.**

Isso ocorre porque o método Random Search pode fornecer uma melhor **exploração do espaço de pesquisa** do que a pesquisa Bayesian optimization no **início**, mas à medida que a quantidade de dados históricos coletados aumenta, essa informação pode ser explorada para beneficiar o processo de pesquisa.

Embora o método pesquisa Bayesian optimization supere a pesquisa Random Search neste caso, nem sempre é esse o caso na prática, especialmente quando o espaço de pesquisa é pequeno e há muitos hiperparâmetros categóricos e condicionais. Devemos selecionar e ajustar diferentes métodos de pesquisa, levando em conta o tamanho do espaço de pesquisa, o número de iterações de pesquisa e as restrições de tempo e recursos (o método Bayesian optimization é executado muito mais lentamente do que o método Random Search). Se nenhuma restrição específica for especificada, a pesquisa Bayesian optimization é um bom lugar para começar.

É importarte ressaltar que os autores explicam o aumento da complexidade que é outro problema que você pode enfrentar ao aplicar a Bayesian optimization é a questão ótima local. Embora tenhamos tentado explorar vários pontos de inicialização ao otimizar a função de aquisição para amostragem, ainda é provável que sempre seja amostrado de uma região local se o modelo substituto não for bem ajustado com base nas amostras históricas ou se a função de aquisição favorecer demais a exploração. Fazer isso levará à exploração a concentrar dem uma região local, ignorando a exploração de outras áreas se a superfície de desempenho da avaliação não for convexa. Além de aumentar a preferência de exploração da função de aquisição, como por exemplo, reduzindo o parâmetro na função de aquisição de UCB, temos os dois truques comumente usados a seguir para evitar que a Bayesian optimization convirja paraum ótimo local:

> Conduz a Bayesian optimization várias vezes e use diferentes sementes aleatórias para amostrar diferentes pontos aleatórios para ajustar o modelo substituto inicial.

> Combine a Bayesian optimization e a Random Search de maneira dinâmica: conduza uma iteração da Random Search após cada várias iterações (digamos, cinco) da pesquisa  Bayesian optimization , alternando as duas.


Além do mais, se o número de iterações de pesquisa for grande e você tiver tempo, é um bom hábito usar a validação cruzada para cada modelo descoberto, em vez de simplesmente avaliar cada modelo em um conjunto de validação fixo.

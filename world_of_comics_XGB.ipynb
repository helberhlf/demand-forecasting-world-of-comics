### Objectives and Key Results (OKRs)


De acordo (SALIM.I; MICHAEL.S; YURI.V, 2018) eles afirmam que muitas **Organizações Exponeciais (ExOs)** estão adotando o método dos **Objetivos e Resultados-Chave (OKR – Objectives and Key Results).** Inventado na Intel pelo CEO Andy Grove e levado ao Google pelo capitalista de risco John Doerr, em 1999, os OKRs acompanham as metas individuais, das equipes e da empresa e expõem os resultados de uma forma aberta e transparente. Em High Output Management, o manual altamente conceituado de Grove, ele introduziu os OKRs como a resposta a duas perguntas simples.

- 1. Aonde quero ir? (*Objetivos.*)
- 2. Como sei que estou chegando lá? (*Resultados-chave para garantir que o progresso esteja ocorrendo.*)

Além da Intel e do Google, outras empresas de rápido crescimento que utilizam o sistema incluem LinkedIn, Zynga, Oracle, Twitter e Facebook.

Além do mais, os autores ressaltar que um programa de OKRs, como o próprio nome sugere, opera ao longo
de dois trajetos. **Um objetivo, por exemplo, poderia ser "aumentar as vendas em 25%", juntamente com "formar duas parcerias estratégicas" e "conduzir uma campanha de AdWords como sendo os resultados-chave
desejados. Os OKRs envolvem foco, simplicidade, ciclos de feedback (mais) curtos e transparência. Como resultado, insights e melhorias são mais fáceis de ver e implementar. Em contraste, a complexidade, o sigilo e objetivos amplos tendem a impedir o progresso, muitas vezes resultando
em consequências imprevisíveis.**

Além disso, os OKRs também são indicadores importantes para as corporações, embora os OKRs sejam mais importantes em novas startups, onde altas taxas de crescimento de contratações exigem um ciclo de feedback mais curto. Mas as grandes empresas também precisam deles porque os OKRs:

- Incentivam o pensamento disciplinado (grandes metas virão à tona);

- Aumentam a comunicação eficaz (todos aprendem o que é importante);

- Estabelecem indicadores para medir o progresso (mostram em que ponto a empresa se encontra);

- Focalizam no esforço (e, assim, sincronizam a organização). Incentivam o pensamento disciplinado (grandes metas virão à tona);

### 1. The end-to-end pipeline of an Machine Learning project

### 1.1. An overview of the end-to-end pipeline

Um pipeline de ML é uma sequência de etapas para conduzir um projeto de ML. Essas etapas seguem:

- **Definicão do problema e coleta de dados:** Define objetivo do problema em termo de negócios como um problema de ML e colete os dados de que você precisa.

- **Pré-processamento de dados e engenharia de recursos:** Processar os dados em um formato adequado que
pode ser inserido nos algoritmos de ML.Assim como,selecionar ou gerar recursos relacionados
para a saída de destino para melhorar o desempenho dos algoritmos.Além disso, essa etapa geralmente é realizada explorando primeiro o conjunto de dados para ter uma noção de suas características. As operações devem acomodar os algoritmos de ML específicos que você está considerando.

- **Seleção de algoritmo de ML:** Selecione algoritmos de ML apropriados para a tarefa que você
gostaria de testar, com base em seu conhecimento prévio do problema e em sua
experiência.

- **Treinamento e avaliação do modelo:** Aplique o algoritmo (ou algoritmos) de ML selecionado
para treinar um modelo de ML com seus dados de treinamento e avaliar seu desempenho em
o conjunto de dados de validação.

- **Ajuste de hiperparâmetros:** Tente alcançar um melhor desempenho iterativamente
ajustando os hiperparâmetros do modelo.

- **Implantação de serviços e monitoramento de modelos:** Implante a solução final de ML e monitore seu desempenho para que você possa manter e melhorar o pipeline continuamente.

Dessa forma, como resultado você pode ver, um projeto de ML é um processo humano em loop. Bem como, iniciando com definnição do problema e coleta de dados, assim o pipeline envolve várias etapas de processamento de dados, que normalmente acontecem de forma assíncrona conforme a figura a seguir. 



> Source: (QINGQUAN.S; HAIFENG.J; X.HU, 2022) 

A primeira coisa que você precisa fazer em qualquer projeto de ML é define o problema e coletar os dados correspondentes.Assim, a definição do problema exige que você especifique as entradas e saídas do modelo de ML. No problema de vendas de video games, as entradas são conjunto de características que descrevem os jogos. ### 2.2. Feature engineering

A Engenharia de Atributos é a 'arte' de formular recursos úteis a partir de 
dados existentes seguindo o destino a ser aprendido e o modelo de aprendizado 
de máquina a ser usado. 

Envolve transformar dados em formas que melhor se relacionem com o alvo 
a ser aprendido. Quando bem feita, a Engenharia de Atributos pode aumentar o 
valor dos dados existentes e melhorar o desempenho dos modelos de Machine 
Learning. Por outro lado, usar recursos ruins pode exigir que você construa 
modelos muito mais complexos para atingir o mesmo nível de desempenho.


Essa é a razão pela qual a Engenharia de Atributos encontrou seu lugar como

### 3.2. Algoritmo XGBoost Feature Importance

**Parâmetro:**  importance_type

- **weight:** o número de vezes que um recurso (variável) é usado para dividir os dados em todas as árvores.

- **gain:** o ganho médio em todas as divisões em que o recurso é usado.

- **cover:** a cobertura média em todas as divisões em que o recurso é usado.

- **total_gain:** o ganho total em todas as divisões em que o recurso é usado.

- **total_cover:** a cobertura total em todas as divisões em que o recurso é usado.

A implementação da função a seguir, seleciona the scores das melhores variáveis para treinar o modelo.

```python
# function call Feature Importance
xbg_best_features = dl_and_ml.feature_imp(features=xFeatures,
                      target=y,
                      param_imp='weight',
                      n_best_features=8)
# Capturando k melhores variáveis para treinamento do modelo
xbg_best_features
```

### 6.1.2. Hiperparameters


> O que significa ajuste de hiperparâmetro?

Segundo ((TOUSHIK) defende que no aprendizado de máquina, **a otimização ou ajuste de hiperparâmetros é o problema de escolher um conjunto de hiperparâmetros ideais para um algoritmo de aprendizado.** Um hiperparâmetro é um parâmetro cujo valor é usado para controlar o processo de aprendizado, ou seja, é uma configuração externa ao modelo e cujo valor não pode ser estimado a partir de dados. Por outro lado, os valores de outros parâmetros (normalmente os pesos dos nós) são aprendidos. O processo de ajuste de hiperparâmetro é uma caminhada na corda bamba para alcançar um equilíbrio entre **underfitting e overfitting**. Underfitting é quando o modelo de aprendizado de máquina não consegue reduzir o erro para o conjunto de teste ou treinamento.

--- 
> Suas principais características são:

- Eles são frequentemente usados em processos para ajudar a estimar os parâmetros do modelo.

- Eles são frequentemente especificados pelo pesquisador. 

- Eles geralmente podem ser configurados usando heurísticas.

- Eles são frequentemente ajustados para um determinado problema de modelagem preditiva.

---

> O que é exemplo de ajuste de hiperparâmetro?

Observa-se que (TOUSHIK) ressaltar alguns exemplos de hiperparâmetros de modelo incluem: A penalidade na regularização do Classificador de Regressão Logística, ou seja, L1 ou L2. A taxa de aprendizado para treinar uma rede neural. Os hiperpârmetros C e sigma para máquinas vetoriais de suporte.


> Por que usamos o ajuste de hiperparâmetros?

Note-se,(TOUSHIK) refina a importância do ajuste de hiperparâmetros é uma parte essencial do controle do comportamento de um modelo de aprendizado de máquina. Se não ajustarmos corretamente nossos hiperparâmetros, nossos parâmetros de modelo estimados produzirão resultados abaixo do ideal, pois não minimizarão a função de perda. **Isso significa que nosso modelo comete mais erros.**

--- 
Dessa forma, não temos como conhecer o melhor valor para um hiperparâmetro em um determinado problema. Podemos usar regras básicas, copiar valores usados em outros problemas ou  procurar o melhor valor por tentativa e erro. Quando um algoritmo de aprendizagem de máquina é ajustado para um problema  específico, usando por exemplo **grid search ou random search**, você está ajustando os hiperparâmetros do modelo para descobrir os parâmetros do modelo que resultam nas previsões mais precisas.Assim acresce-se que, quando se trata de otimizar um modelo, você testa diversas variações de hiperparâmetros nesse modelo com todas as combinações possíveis de hiperparâmetros em um espaço de busca.


Em poucas palavras, hiperpârametros se refere a números que você usa para acertar ou ajustar o desempenho de um modelo estatístico ou de aprendizado de máquina.

#### 2. silent [default=0]

- **Ativar o modo silencioso silencioso, defina-o para 1, ou seja, desligado; assim, nenhuma mensagem será impressa.** Geralmente, é uma boa ideia mantê-lo 0, pois as mensagens podem ajudar a entender o modelo; e como as métricas estão indo.

### <center>Booster parameters</center>

Então esta parte descata-se a qual reforço você escolhará. Temos a seguintes propriedades sobre boosters baseados em árvores destacam-se:

#### 1. eta [default=0.3]

- Semelhante ao learning_rate; trabalha com generalização.
- Valores finais típicos a serem usados: 0,01-0,2

Segundo (ANDREW e BRUCE, 2019) o fator de encolhimento ou taxa de aprendizagem está entre 0 e 1 aplicado a α no algoritmo de boosting. O padrão é 0,3, **mas, para dados ruidosos, recomenda-se valores menores (por exemplo, 0,1)**. Nesse mesmo sentido, (HARRIOSON ,2020) apura que após cada passo de boosting, (melhoria) os pesos recém-adicionados são escalados de acordo com esse fator. Quanto menor o valor, mais conservador será, mas também serão necessárias mais árvores para convergir. Na chamada ao dados de train, você pode passar um parâmetro learning_rates, que é uma lista de taxas em cada rodada (isto é, [.1] * 100 + [.05] * 100)


#### 2. min_child_weight [default=1]

- Define **a soma mínima dos pesos de todas as observações necessárias em um filho**. É usado para **controlar o ajuste excessivo (overfitting)**. Valores mais altos impedem que um modelo aprenda  relações que podem ser altamente específicas para a amostra específica selecionada para uma árvore.


- Valores muito altos podem levar a um ajuste insuficiente, portanto, ele deve ser ajustado usando CV.


#### 3. max_depth [padrão=6]

- A **profundidade máxima** de uma árvore também é usada para **controlar o ajuste excessivo**, pois uma profundidade maior permitirá que o modelo aprenda relações muito específicas para uma amostra específica. Ou seja, ao contrário da floresta aleatória **(random forest)**, que ajusta árvores muito profundas, o **boosting costuma ajustar árvores rasas**. Isso tem a vantagem de evitar interações falsas e complexas no modelo, as quais podem ser resultado de dados ruidosos (ANDREW e BRUCE, 2019).


- Deve ser ajustado usando CV. Valores típicos: 3-10


#### 4. max_leaf_nodes

- O número máximo de nós terminais ou folhas em uma árvore.

- Como as árvores binárias são criadas, uma profundidade de 'n' produziria um máximo de 2^n folhas, ou seja, em que  cada nó da árvore tem  no máximo 2 filhos. Se isso for definido, o GBM ignorará max_depth.


#### 5. gamma [default=0]

- Um nó é dividido apenas quando a divisão resultante fornece uma redução positiva na função de perda.**Gamma especifica a redução de perda mínima necessária para que ocorra uma divisão.**

- Torna o algoritmo conservador. Os valores podem variar dependendo da função de perda e devem ser ajustados. Pode ser usado para controlar o overfitting.

Observa-se (HARRISON, 2020) refina que gama controla a poda (pruning). Assim como, pode varia de 0 a infinito, a redução de perda mínima necessária para separar mais uma folha. Quanto maior o valor de gama, mais conservador será. Se as pontuações de treinamento e de teste divergirem, insira um número maior (em torno de 10). Se scores de treinamento e teste estiverem próximas, utilize um número menor.

#### 6. max_delta_step [default=0]

- No passo **delta máximo, permitimos que a estimativa de peso de cada árvore seja**. Se o valor for definido como 0, significa que não há restrição. Se estiver definido com um valor positivo, pode ajudar a tornar a etapa de atualização mais conservadora.

- Isso geralmente não é usado, mas sim, usado para classes desbalanceadas, pois deixa as atualizações mais conservadoras com valores de 1 a 10.


#### 7. subsample [default=1]

- Denota a fração de observações a serem amostras aleatórias para cada árvore.

- Valores mais baixos tornam o **algoritmo mais conservador e evitam overfitting**, mas valores muito pequenos podem levar a underfitting.

- Valores típicos: 0. 5 - 1

- Conforme discutido, a subsample especifica a fração de instâncias de treinamento a serem usadas para treinar cada árvore. Por exemplo, se subsample=0.75, cada árvore é treinada em **75% das instâncias de treinamento selecionadas aleatoriamente.** Como você provavelmente já deve ter adivinhado, essa técnica troca um **viés maior por uma variância menor.** Ela também acelera substancialmente o treinamento. Essa técnica se chama **gradiente estocástico boosting.**



#### 8. colsample_bytree [padrão=1]

- Mostra a fração de colunas a serem amostras aleatórias para cada árvore. Colsample_bytree menor fornece regulação adicional.

- Valores típicos: 0.5 -1

#### 9. colsample_bylevel [default=1]

- Indica a proporção de **subamostra de colunas para cada divisão, em cada nível**.

#### 10. colsample_bynode [default=1]

- Indica a proporção de **subamostra de colunas para cada divisão, em cada nó**.


#### 11. alpha [default=0]


- **Termo  de regularização L1.**

- A regularização L1 força os pesos dos recursos não informativos a serem zero, subtraindo uma pequena quantidade do peso em cada iteração e, assim, tornando o peso zero, eventualmente. Também é chamado de regularização por esparsidade <a href="https://developers-google-com.translate.goog/machine-learning/crash-course/regularization-for-sparsity/l1-regularization?_x_tr_sl=en&_x_tr_tl=pt&_x_tr_hl=pt-BR&_x_tr_pto=sc">regularização por esparsidade</a>. Aplica-se a pesos de folha (em vez de pesos de recursos); valor maior significa mais regularização


#### 12. lambda [default=1]

- **Termo de regularização L2.**

- Isso costumava lidar com a parte de regularização do XGBoost. Deve ser explorado para reduzir o overfitting

- A regularização L2 age como uma força que remove uma pequena porcentagem de pesos a cada iteração. Portanto, os pesos nunca serão iguais a zero. A regularização L2 penaliza (weight)². Existe um parâmetro adicional para ajustar o termo  de regularização L2 que é chamado de taxa de regularização <a href="https://developers-google-com.translate.goog/machine-learning/crash-course/regularization-for-simplicity/lambda?_x_tr_sl=en&_x_tr_tl=pt&_x_tr_hl=pt-BR&_x_tr_pto=sc">(lambda)</a>.

- Além do mais, é mais suave que alfa. Assim sendo, aplica-se aos pesos das folhas.


> *A regularização é um jeito de evitar o overfitting através da inclusão de um termo de penalidade no número de parâmetros (por exemplo, tamanho da árvore) em um modelo.*


#### 12. scale_pos_weight [default=1]

### 6.9. Ajustando os Parâmetros de Regularização (alfa e lambda)

### 6.9.1. L1 vs L2 Regularização

De acordo (DEVELOPERS-GOOGLE) esclarece alguns fundamentos específicos sobre as regulariação L2 e L1 que apresentadas da seguintes formas:


L2 e L1 penalizam os pesos de forma diferente:

- L2 penaliza weight².
- L1 penaliza |weight|.

Consequentemente, L2 e L1 têm derivadas diferentes:

- A derivada de L2 é **2 * weight**

A derivada de L1 é **k (uma constante, cujo valor é independente do peso).**

Nesssa mesma linha de trabalho, (DEVELOPERS-GOOGLE) defende que você pode pensar na **derivada de L2 como uma força que remove x% do peso todas as vezes**. DEVELOPERS-GOOGLE ressaltar um conceito importante por meio das palavras do autor **ZENO** de acordo com ele *mesmo se você remover x por cento de um número bilhões de vezes, o número diminuído ainda nunca chegará a zero.* (Zeno estava menos familiarizado com as limitações de precisão de ponto flutuante, que poderiam produzir exatamente zero.) De qualquer forma, **L2 normalmente não leva os pesos a zero.**

Conseqüentemente, você pode pensar na **derivada de L1 como uma força que subtrai alguma constante do peso todas as vezes**. No entanto, graças aos **valores absolutos, L1 tem uma descontinuidade em 0, o que faz com que os resultados de subtração que cruzam 0 sejam zerados**. Por exemplo, se a subtração forçaria um peso de +0,1 a -0,2, L 1 definirá o peso exatamente 0. <a href ="https://pt.wikipedia.org/wiki/Eureka_(exclama%C3%A7%C3%A3o)">Eureka</a>, L1 zerou o peso.

>A regularização L 1  penalizando o valor absoluto de todos os pesos, acaba sendo bastante eficiente para modelos largos.

```python
# calculate confidence interval for the RMSE
def ci_rmse(y_pred, y_test, confidence = .95):

    squared_errors = (y_pred - y_test) ** 2
    return np.sqrt(stats.t.interval(confidence, len(squared_errors) - 1,
                         loc=squared_errors.mean(),scale=stats.sem(squared_errors)))

# function call 
ci_rmse(reg_XGB.predict(X_test_min_max), y_test)
```
